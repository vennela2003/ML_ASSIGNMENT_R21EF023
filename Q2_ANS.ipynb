```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
preprocessed_data = pd.read_csv('50_Startups.csv')
```

```python
# Plot 1: Profit Distribution
plt.figure(figsize=(12, 6))
sns.histplot(data=preprocessed_data, x="Profit", kde=True)
plt.title("Profit Distribution")
plt.xlabel("Profit")
plt.ylabel("Frequency")
plt.show()
```

```python
# Plot 2: R&D Spend vs. Profit
plt.figure(figsize=(12, 6))
sns.scatterplot(data=preprocessed_data, x="R&D Spend", y="Profit")
plt.title("R&D Spend vs. Profit")
plt.xlabel("R&D Spend")
plt.ylabel("Profit")
plt.show()
```

```python
# Plot 3: Marketing Spend vs. Profit
plt.figure(figsize=(12, 6))
sns.scatterplot(data=preprocessed_data, x="Marketing Spend", y="Profit")
plt.title("Marketing Spend vs. Profit")
plt.xlabel("Marketing Spend")
plt.ylabel("Profit")
plt.show()
```

```python
# Plot 4: Administration vs. Profit
plt.figure(figsize=(12, 6))
sns.scatterplot(data=preprocessed_data, x="Administration", y="Profit")
plt.title("Administration vs. Profit")
plt.xlabel("Administration")
plt.ylabel("Profit")
plt.show()
```

```python
# Plot 5: Pairplot for all numerical variables
sns.pairplot(preprocessed_data)
plt.show()
```

```python
# Plot 6: Correlation Heatmap
correlation_matrix = preprocessed_data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Plot 7: Profit vs. R&D Spend with regression line by State
plt.figure(figsize=(12, 6))
sns.lmplot(data=preprocessed_data, x="R&D Spend", y="Profit", height=6)
plt.title("Profit vs. R&D Spend with Regression Line by State")
plt.xlabel("R&D Spend")
plt.ylabel("Profit")
plt.show()
```

```python
# Plot 8: Pairplot with Hue by Marketing Spend Range
preprocessed_data["Marketing Spend Range"] = pd.cut(preprocessed_data["Marketing Spend"], bins=3)
plt.figure(figsize=(12, 6))
sns.pairplot(preprocessed_data, hue="Marketing Spend Range")
plt.title("Pairplot with Hue by Marketing Spend Range")
plt.show()
```

```python
# Plot 9: Boxplot of Profit by Company Region
preprocessed_data["Company Region"] = preprocessed_data["State_California"].apply(lambda x: "California" if x == "California" else "Non-California")
plt.figure(figsize=(12, 6))
sns.boxplot(data=preprocessed_data, x="Company Region", y="Profit")
plt.title("Boxplot of Profit by Company Region")
plt.xlabel("Company Region")
plt.ylabel("Profit")
plt.show()
```

```python
# Plot 10: Scatterplot of R&D Spend vs. Marketing Spend by Profit
plt.figure(figsize=(12, 6))
scatterplot = sns.scatterplot(data=preprocessed_data, x="R&D Spend", y="Marketing Spend", hue="Profit", palette="viridis")
plt.title("Scatterplot of R&D Spend vs. Marketing Spend by Profit")
plt.xlabel("R&D Spend")
plt.ylabel("Marketing Spend")

# Customize the legend
scatterplot.legend(title="Profit")
plt.show()
```
